# Jumlah Usaha Tani Per Pengguna Lahan (Tahun 2003 dan 2013)

Berikut ini merupakan program analisa data sederhana menggunakan python dan library pandas.
Analisa berupa statistik tentang data dan visualisasi sederhana dari data.

## Derkripsi Data

```python
"""
This is exercise to analyze data from dataset csv

By Aditio Agung Nugroho
"""

import pandas as pd
import os

print('This is a dataset which contains about number of farmer business based on type of land used.\nThis data wes gathered between 2003 and 2013.')
data_farmer_type = pd.read_csv('processed-jumlah-usaha-tani-per-pengguna-lahan-gurem-2003-2013.csv',
                              delimiter = ',', encoding = "ISO-8859-1")
data_farmer_type
```

## Statistik Data

```python
# Sum all farmers
print('Count all farmers:', data_farmer_type.jumlah.sum())
# Sum all farmers in 2003
print('Count all farmers in 2003:', data_farmer_type[data_farmer_type['tahun'] == 2003].jumlah.sum())
# Sum all farmers in 2013
print('Count all farmers in 2013:', data_farmer_type[data_farmer_type['tahun'] == 2013].jumlah.sum(), '\n')

# Sum all farmers which use field
field_condition = data_farmer_type['jenis_pengguna'] == 'Pengguna Lahan'
print('Count all field farmer:', data_farmer_type[field_condition].sum().jumlah)
# Sum all farmers which use field in 2003
year_condition = data_farmer_type['tahun'] == 2003
print('Count all field farmer in 2003:', data_farmer_type[field_condition & year_condition].sum().jumlah)
# Sum all farmers which use field in 2013
year_condition = data_farmer_type['tahun'] == 2013
print('Count all field farmer in 2013:', data_farmer_type[field_condition & year_condition].sum().jumlah, '\n')

# Sum all farmers which use small farm
field_condition = data_farmer_type['jenis_pengguna'] == 'Pertanian Gurem'
print('Count all small farm farmer:', data_farmer_type[field_condition].sum().jumlah)
# Sum all farmers which use field in 2003
year_condition = data_farmer_type['tahun'] == 2003
print('Count all small farm farmer in 2003:', data_farmer_type[field_condition & year_condition].sum().jumlah)
# Sum all farmers which use field in 2013
year_condition = data_farmer_type['tahun'] == 2013
print('Count all small farm farmer in 2013:', data_farmer_type[field_condition & year_condition].sum().jumlah)
```

```python
# Smallest number of farmer in one province
print('Smallest farmers number in one province:', data_farmer_type.jumlah.min())
# Smallest number of farmer in one province in 2003
year_condition = data_farmer_type['tahun'] == 2003
print('Smallest farmers number in one province in 2003:', data_farmer_type[year_condition].jumlah.min())
# Smallest number of farmer in one province in 2013
year_condition = data_farmer_type['tahun'] == 2013
print('Smallest farmers number in one province in 2013:', data_farmer_type[year_condition].jumlah.min(), '\n')

# Biggest number of farmer in one province
print('Biggest farmers number in one province:', data_farmer_type.jumlah.max())
# Biggest number of farmer in one province in 2003
year_condition = data_farmer_type['tahun'] == 2003
print('Biggest farmers number in one province in 2003:', data_farmer_type[year_condition].jumlah.max())
# Biggest number of farmer in one province in 2013
year_condition = data_farmer_type['tahun'] == 2013
print('Biggest farmers number in one province in 2013:', data_farmer_type[year_condition].jumlah.max(), '\n')

# Mean of farmers using land and small farm
print('Mean of farmers using land and small farm:', data_farmer_type.jumlah.mean())
# Mean of farmers using land and small farm in 2003
year_condition = data_farmer_type['tahun'] == 2003
print('Mean of farmers using land and small farm in 2003:', data_farmer_type[year_condition].jumlah.mean())
# Mean of farmers using land and small farm in 2013
year_condition = data_farmer_type['tahun'] == 2013
print('Mean of farmers using land and small farm in 2013:', data_farmer_type[year_condition].jumlah.mean(), '\n')

# Median of farmers using land and small farm
print('Median of farmers using land and small farm:', data_farmer_type.jumlah.median())
# Median of farmers using land and small farm in 2003
year_condition = data_farmer_type['tahun'] == 2003
print('Median of farmers using land and small farm in 2003:', data_farmer_type[year_condition].jumlah.median())
# Median of farmers using land and small farm in 2013
year_condition = data_farmer_type['tahun'] == 2013
print('Median of farmers using land and small farm in 2013:', data_farmer_type[year_condition].jumlah.median())
```

```python
# Total farmers group by province, type field, and year
data_farmer_type.groupby(['nama_provinsi', 'jenis_pengguna', 'tahun', 'jumlah']).sum()
```

## Visualisasi Data

```python
# Total farmers visualization
ax = data_farmer_type.groupby('jenis_pengguna').sum()['jumlah']
ax = ax.plot(kind='bar', figsize=(10,4), title="Jumlah Berdasarkan Jenis Lahan Tahun 2003-2013", legend=True, fontsize=12)
ax.set_xlabel("Jenis Pengguna", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

Berdasarkan total data petani di tahun 2003 dan 2013 dapat dilihat bahwa petani yang menggunakan lahan lebih banyak dibanding petani yang menggunakan pertanian gurem

```python
# Total farmers visualization in 2003
year_condition = data_farmer_type['tahun'] == 2003
ax = data_farmer_type[year_condition].groupby('jenis_pengguna').sum()['jumlah']
ax = ax.plot(kind='bar', figsize=(10,4), title="Jumlah Berdasarkan Jenis Lahan Tahun 2003", legend=True, fontsize=12)
ax.set_xlabel("Jenis Pengguna", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
# Total farmers visualization in 2013
year_condition = data_farmer_type['tahun'] == 2013
ax = data_farmer_type[year_condition].groupby('jenis_pengguna').sum()['jumlah']
ax = ax.plot(kind='bar', figsize=(10,4), title="Jumlah Berdasarkan Jenis Lahan Tahun 2013", legend=True, fontsize=12)
ax.set_xlabel("Jenis Pengguna", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

Berdasarkan visualisasi data di atas dapat dilihat bahwa petani yang menggunakan lahan tetap lebih banyak dari petani yang menggunakan pertanian gurem baik di tahun 2003 maupun di tahun 2013
