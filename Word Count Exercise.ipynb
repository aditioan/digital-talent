# Word Count Program

This is a simple program to count total word appearance in a paragraph

```python
"""
This is exercise to count number of string in paragraph for basic python

By Aditio Agung Nugroho
"""

try:
    raw_input           # Python 2
except NameError:
    raw_input = input   # Python 3


if __name__ == '__main__':
    paragraph = raw_input('Enter a paragraph:\n').replace('.', '').replace(',', '')
    list_words = paragraph.split(' ')
    unique_word = []
    for i in list_words:
        if i not in unique_word:
            unique_word.append(i)
            print(i, ':', list_words.count(i), '\n')
```
