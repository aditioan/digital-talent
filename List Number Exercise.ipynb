# List Number Program

This is exercise to find odd number, count condition, stardard deviation, maximum, minimum, and mode of a random list number 0-100 in 250 items

```python
"""
This is exercise list number for basic python

By Aditio Agung Nugroho
"""
import random
import math

try:
    xrange           # Python 2
except NameError:
    xrange = range   # Python 3


def odd_number(data):
    """Select odd number from random list data

        :param data: Random list data
        :return: List of odd number in data

        Examples:
        >>> odd_number([71, 73, 22])
        [71, 73]

        >>> odd_number([65, 70, 12])
        [65]
    """
    odd_data = []
    for i in data:
        if i % 2 != 0:
            odd_data.append(i)
    return odd_data


def count_condition(data, number):
    """Count how many data which have a value more than number

        :param number: Number threshold
        :return: Length of the data more than threshold

        Examples:
        >>> count_condition([71, 73, 22], 70)
        2

        >>> count_condition([65, 70, 12], 65)
        1
    """
    condition_list = []
    for i in data:
        if i > number:
            condition_list.append(i)
    return len(condition_list)


def standar_dev(data):
    """Count standard deviation from a list data

        :param data: Random list of number
        :return: Standard deviation of the list number

        Examples:
        >>> standar_dev([172, 167, 180, 170, 169, 160, 175, 165, 173, 170])
        5.51
    """
    x_total = 0
    x_power = 0
    for i in data:
        x_total += i
        x_power += i ** 2
    sdev = round(math.sqrt((len(data) * x_power - x_total ** 2) / (len(data) * (len(data) - 1))), 2)

    return sdev


def mode_data(data):
    """Search mode of a list data

        :param data: Random list of number
        :return: Mode of the list number

        Examples:
        >>> mode_data([172, 167, 180, 170, 169, 160, 175, 165, 173, 170])
        170
    """
    data_appearance = 0
    for i in data:
        if data.count(i) > data_appearance:
            data_appearance = data.count(i)
            mode = i

    return mode


if __name__ == '__main__':
    data_random = [random.choice(range(100)) for i in range(250)]
    print('Current random list data are:\n', data_random)
    data_odd = odd_number(data_random)
    print('Odd number in the current data are:\n', data_odd)
    data_condition = count_condition(data_random, 70)
    print('Count of the number which more than 70 in the current data is:\n', data_condition)
    data_sdev = standar_dev(data_random)
    print('Standard deviation of the current data is:\n', data_sdev)
    print('Maximum number of the current data is:\n', max(data_random))
    print('Minimum number of the current data is:\n', min(data_random))
    data_mode = mode_data(data_random)
    print('Mode number of the current data is:\n', data_mode)
```
